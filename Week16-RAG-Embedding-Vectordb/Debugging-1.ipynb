```python
import faiss
import numpy as np

d = 128
index = faiss.IndexFlatL2(d)

np.random.seed(42)

# Database vectors
data = np.random.rand(10, d).astype(np.float32)
index.add(data)

# Query vector MUST be 2D and float32
query = np.random.rand(d).astype(np.float32)
query = np.expand_dims(query, axis=0)   # shape becomes (1, 128)

# Search
D, I = index.search(query, k=3)

print("Nearest neighbors:", I)
print("Distances:", D)
```

```python
import google.generativeai as genai
import numpy as np
import faiss

genai.configure(api_key="YOUR_API_KEY")

# Step 1: Create documents
documents = [
    "Artificial Intelligence is the future",
    "Python is a programming language",
    "FAISS is used for similarity search"
]

# Step 2: Create embeddings
doc_embeddings = []

for doc in documents:
    result = genai.embed_content(
        model="models/embedding-001",
        content=doc,
        task_type="retrieval_document"
    )
    doc_embeddings.append(result["embedding"])

doc_embeddings = np.array(doc_embeddings).astype("float32")

# Step 3: Create FAISS index
d = len(doc_embeddings[0])
index = faiss.IndexFlatL2(d)
index.add(doc_embeddings)

# Step 4: Embed query (IMPORTANT FIX)
query = "What is AI?"

query_result = genai.embed_content(
    model="models/embedding-001",
    content=query,
    task_type="retrieval_query"
)

query_embedding = np.array(query_result["embedding"]).astype("float32")
query_embedding = np.expand_dims(query_embedding, axis=0)

# Step 5: Search
D, I = index.search(query_embedding, k=1)

print("Best matching document:")
print(documents[I[0][0]])
```
