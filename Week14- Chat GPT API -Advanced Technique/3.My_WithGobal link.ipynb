```python
import gradio as gr
from google import genai

# Initialize Gemini client
client = genai.Client(api_key="YOUR_GEMINI_API_KEY")

# Store chat history
chat_history = []

# System instruction
SYSTEM_PROMPT = """
You are a real estate financial expert chatbot.
You help users with:
- Property investment advice
- ROI calculations
- Rental income analysis
- Real estate financial planning
Keep answers clear and professional.
"""

def chatbot(user_message, history):

    # Convert history to Gemini format
    contents = []

    # Add system instruction first
    contents.append(SYSTEM_PROMPT)

    # Add previous conversation
    for user, bot in history:
        contents.append(f"User: {user}")
        contents.append(f"Assistant: {bot}")

    # Add new message
    contents.append(f"User: {user_message}")

    # Generate response
    response = client.models.generate_content(
        model="gemini-1.5-flash",  # Free model
        contents=contents,
        config={
            "temperature": 0.7,
            "max_output_tokens": 500,
        }
    )

    bot_reply = response.text

    history.append((user_message, bot_reply))

    return history, history


# Gradio interface
demo = gr.Interface(
    fn=chatbot,
    inputs=[
        gr.Textbox(label="Ask Real Estate Question"),
        gr.State([])
    ],
    outputs=[
        gr.Chatbot(label="Real Estate Financial Advisor"),
        gr.State([])
    ],
    title="Real Estate Financial Chatbot (Gemini)",
    description="Ask about property investment, ROI, rental income, etc."
)

# Launch app
demo.launch(share=True)
```
