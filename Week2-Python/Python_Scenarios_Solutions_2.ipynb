## Question 1: Identifying Transaction Type

# Explanation
This program identifies whether a transaction is Deposit, Withdrawal, or No Transaction.

# Code

```python
num = float(input("Enter transaction amount: "))

if num > 0:
    print("Positive (Deposit)")
elif num < 0:
    print("Negative (Withdrawal)")
else:
    print("Zero (No Transaction)")
```

## Question 2: Sum of Digits

# Explanation
This program calculates the sum of digits of a number.

# Code

```python
number = input("Enter a number: ")
digit_sum = 0

for digit in number:
    digit_sum += int(digit)

print("Sum of digits:", digit_sum)
```

## Question 4: Prime Number Check

# Explanation
This program checks whether a number is prime.

# Code

```python
import math

num = int(input("Enter number: "))

if num < 2:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime")
    else:
        print("Not Prime")
```

## Question 5: Factorial Using Recursion

# Explanation
This program calculates factorial using recursion.

# Code

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

num = int(input("Enter number: "))
print("Factorial:", factorial(num))
```

## Question 6: Armstrong Number Check

# Explanation
This program checks whether a number is an Armstrong number.

# Code

```python
num = input("Enter number: ")
power = len(num)
total = 0

for digit in num:
    total += int(digit) ** power

if total == int(num):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

## Question 7: Swap First and Last Characters

# Explanation
This program swaps the first and last characters of a string.

# Code

```python
text = input("Enter string: ")

if len(text) < 2:
    print(text)
else:
    swapped = text[-1] + text[1:-1] + text[0]
    print("Modified string:", swapped)
```

## Question 8: Decimal to Binary Conversion

# Explanation
This program converts decimal number to binary.

# Code

```python
num = int(input("Enter decimal number: "))

binary = ""

while num > 0:
    binary += str(num % 2)
    num //= 2

binary = binary[::-1]

print("Binary representation:", binary)
```

## Question 9: Longest Word in Sentence

# Explanation
This program finds the longest word in a sentence.

# Code

```python
sentence = input("Enter sentence: ")
words = sentence.split()

longest = words[0]

for word in words:
    if len(word) > len(longest):
        longest = word

print("Longest word:", longest)
```

## Question 10: Anagram Check

# Explanation
This program checks whether two strings are anagrams.

# Code

```python
str1 = input("Enter first string: ").replace(" ", "").lower()
str2 = input("Enter second string: ").replace(" ", "").lower()

if sorted(str1) == sorted(str2):
    print("Anagram")
else:
    print("Not an Anagram")
```
