```python
# Question 1 - If Else
i = int(input('value: '))
if i == 10:
    print('Correct')
else:
    print('Wrong')
```

```python
# Question 2
password = input('Enter the password: ')
if password == 'BALA@123':
    print('Your password is correct')
else:
    print('Your password is wrong')
```

```python
# Question 3
age = int(input('age: '))
if age <= 18:
    print('Children')
elif age <= 35:
    print('Adult')
elif age <= 60:
    print('Citizen')
else:
    print('Senior Citizen')
```

```python
# Question 4
n = int(input('Enter any number:'))
if n > 0:
    print('Number is positive')
elif n < 0:
    print('Number is negative')
else:
    print('Zero')
```

```python
# Question 5
n = int(input('Enter a number to check:'))
if n % 5 == 0:
    print('Number is divisible by 5')
else:
    print('Number is not divisible by 5')
```
