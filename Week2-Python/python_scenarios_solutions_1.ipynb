```python
## 1. Check Voting Eligibility
age = int(input("Enter your age: "))
if age >= 18:
    print("Eligible to vote.")
else:
    print("Not eligible to vote.")
```

```python
## 2. Find Largest Number in a List
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest number is:", largest)
```

```python
## 3. Calculate Bonus
salary = float(input("Enter salary: "))

if salary > 50000:
    bonus = salary * 0.10
else:
    bonus = 0

print("Bonus amount:", bonus)
```

```python
## 4. Check Even or Odd
number = int(input("Enter a number: "))

if number % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
## 5. Reverse Word or Sentence
text = input("Enter a word or sentence: ")

char_list = list(text)
char_list.reverse()

reversed_text = ''.join(char_list)

print("Reversed text:", reversed_text)
```

```python
## 6. Pass or Fail System
marks = float(input("Enter student's marks: "))

if marks >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
## 7. Apply Discount
amount = float(input("Enter total order amount: "))

if amount > 100:
    discount = amount * 0.20
else:
    discount = 0

final_amount = amount - discount

print("Final amount to pay:", final_amount)
```

```python
## 8. Bank Withdrawal System
balance = float(input("Enter account balance: "))
withdrawal = float(input("Enter withdrawal amount: "))

if withdrawal <= balance:
    balance -= withdrawal
    print("Withdrawal successful. Updated balance:", balance)
else:
    print("Insufficient funds.")
```

```python
## 9. Leap Year Check
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap year")
elif year % 100 == 0:
    print("Not a leap year")
elif year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```

```python
## 10. Filter Even Numbers from List
numbers = list(map(int, input("Enter numbers separated by space: ").split()))
even_numbers = []

for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)

print("Even numbers:", even_numbers)
```
