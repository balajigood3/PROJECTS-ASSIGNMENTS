```python
class multipleFunctions:
    
    @staticmethod
    def oddEven():
        num = int(input("Enter the number: "))
        if (num % 2) == 1:
            print("Odd Number")
            return "Odd Number"
        else:
            print("Even Number")
            return "Even Number"

    @staticmethod
    def BMI():
        bmi = float(input("Enter the BMI Index: "))
        if bmi < 18.5:
            print("Underweight")
            return "Underweight"
        elif bmi < 24.9:
            print("Normal")
            return "Normal"
        elif bmi < 29.9:
            print("Overweight")
            return "Overweight"
        else:
            print("Very Overweight")
            return "Very Overweight"


# ðŸ”¹ Call the methods here
multipleFunctions.oddEven()
multipleFunctions.BMI()
```
