```python
from langchain.agents import Tool
from tavily import TavilyClient

# Setup Tavily Client
tavily_client = TavilyClient(api_key=os.getenv("TAVILY_API_KEY"))

# Define a search function that Tavily Tool can use
def tavily_search(query: str) -> str:
    results = tavily_client.search(query=query, max_results=5)
    return "\n\n".join([r["content"] for r in results["results"]])

# Create LangChain Tool
tavily_tool = Tool(
    name="Tavily Web Search",
    func=tavily_search,
    description="Use this tool to search the web for fresh info about companies or markets"
)
```

```python
from langchain.agents import initialize_agent, AgentType

agent = initialize_agent(
    tools=[tavily_tool],  # âœ… correct tool
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)
```
