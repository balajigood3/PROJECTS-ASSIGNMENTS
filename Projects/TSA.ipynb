```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
print("Loaded libraries successfully.")
```

```python
#Loading the dataset
df = pd.read_csv('50_Startups.csv')
```

```python
# converting categorical variable to numeric using one-hot encoding
df = pd.get_dummies(df, columns=['State'], drop_first=True)
print("Dataset loaded and preprocessed successfully.")
```

```python
df.columns
```

```python
#Define features and target variable
X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
Y_df = df['Profit']
```

```python
#splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_df, test_size=0.2, random_state=42)
print("Dataset split into training and testing sets successfully.")
```

```python
#Train the model
model = LinearRegression()
model.fit(X_train, Y_train)
print("Model trained successfully.")
```

```python
#PLOTTING THE RESULTS
Y_pred = model.predict(X_test)
plt.figure(figsize=(10, 6))
sns.scatterplot(x=Y_test, y=Y_pred)
plt.xlabel('Actual Profit')
plt.ylabel('Predicted Profit')
plt.title('Actual vs Predicted Profit')
plt.show()
```
