Dataset expected file name: `Placement_Data_Full_Class.csv`

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_csv('Placement_Data_Full_Class.csv')
df.head()
```

## 1. Replace NaN values and justify

```python
# salary has NaN because students not placed
# replace NaN with 0

df['salary'].fillna(0, inplace=True)
df['salary'].isna().sum()
```

Justification: Salary is missing because students are not placed. So replacing with 0 is logical.

## 2. How many are not placed?

```python
not_placed = df[df['status']=='Not Placed']
len(not_placed)
```

## 3. Reason for non-placement

```python
not_placed.describe()
```

## 4. Relation between salary and mba_p

```python
sns.scatterplot(x='mba_p', y='salary', data=df)
plt.show()

df[['mba_p','salary']].corr()
```

## 5. Specialization getting minimum salary

```python
df.groupby('specialisation')['salary'].min()
```

## 6. Count salary above 500000

```python
len(df[df['salary']>500000])
```

## 7. ANOVA between etest_p and mba_p

```python
stats.f_oneway(df['etest_p'], df['mba_p'])
```

## 8. Similarity between Sci&Tech and Mkt&HR salary

```python
group1 = df[df['degree_t']=='Sci&Tech']['salary']
group2 = df[df['specialisation']=='Mkt&HR']['salary']

stats.ttest_ind(group1, group2)
```

## 9. Convert salary to standard normal

```python
df['salary_zscore'] = stats.zscore(df['salary'])
df[['salary','salary_zscore']].head()
```

## 10. Probability density function between 700000 and 900000

```python
mean = df['salary'].mean()
std = df['salary'].std()

prob = stats.norm.cdf(900000, mean, std) - stats.norm.cdf(700000, mean, std)
prob
```

## 11. Similarity between Sci&Tech etest_p and mba_p

```python
group = df[df['degree_t']=='Sci&Tech']
stats.ttest_ind(group['etest_p'], group['mba_p'])
```

## 12. Highest correlation with salary

```python
corr = df.corr(numeric_only=True)['salary']
corr.sort_values(ascending=False)
```

## 13. Useful graph

```python
sns.boxplot(x='specialisation', y='salary', data=df)
plt.show()
```
