## ReviewerApprover (Blog Publishing)

```python
# Incorrect code
from autogen import ConversableAgent

config = {
    "config_list": [{
        "model": "models/gemini-1.5-pro-latest",
        "api_key": "your-api-key",
        "api_type": "google"
    }]
}

reviewer = ConversableAgent(name="ReviewerAgent", llm_config=config)
approver = ConversableAgent(name="ReviewerAgent", llm_config=config)  #  Same name as reviewer

blog_draft = "AI is good. It helps. Many use it."

#  Dictionary-like access to string reply
review_feedback = reviewer.generate_reply(messages=[{"role": "user", "content": f"Review this blog and suggest improvements: {blog_draft}"}])
approval = approver.generate_reply(messages=[{"role": "user", "content": review_feedback["content"]}])

print("Feedback:", review_feedback["content"])
print("Approval:", approval["content"])
```

```python
from autogen import ConversableAgent

config = {
    "config_list": [{
        "model": "models/gemini-1.5-pro-latest",
        "api_key": "AIzaSyBL55yg3Vrf0PKrhRiMKtAUo-Fkr6E71zY",  # Replace with your Gemini API Key
        "api_type": "google"
    }]
}

# Unique agent names
reviewer = ConversableAgent(name="ContentReviewer", llm_config=config)
approver = ConversableAgent(name="FinalApprover", llm_config=config)

blog_draft = "AI is good. It helps. Many use it."

# Step 1: Reviewer improves the content
review_feedback = reviewer.generate_reply(
    messages=[{"role": "user", "content": f"Please review and improve the following blog post:\n\n{blog_draft}"}]
)

# Step 2: Approver decides if it's ready to publish
approval = approver.generate_reply(
    messages=[{"role": "user", "content": f"Is this blog ready to publish? Reply YES or NO and give a short reason:\n\n{review_feedback['content']}"}]
)

#  Extract only the 'content' for clean output
print(" Reviewer Feedback:\n", review_feedback["content"])
print(" Final Approval:\n", approval["content"])
```
