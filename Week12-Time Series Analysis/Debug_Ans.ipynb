# Q1. ValueError in Rolling Mean Calculation

```python
import pandas as pd

df = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
                   'Sales': [100, 120, 130]})

df['7_day_avg'] = df['Sales'].rolling(window=7).mean()
print(df)
```

# Q4. NameError

```python
# Import necessary libraries
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, header=0, parse_dates=['Month'], index_col='Month')

# Fit ARIMA model
model = ARIMA(df['Passengers'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast next 12 time steps
forecast = model_fit.forecast(steps=12)
```
