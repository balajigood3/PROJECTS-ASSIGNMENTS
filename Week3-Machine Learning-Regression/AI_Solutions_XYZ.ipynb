# 2. Email Classification using NLP

```python
# Dummy email dataset
emails = {
    'message': [
        'Where is my order?',
        'Product quality is bad',
        'Shipment delayed',
        'Payment issue',
        'Refund not received'
    ],
    'department': [
        'Shipping',
        'Quality',
        'Shipping',
        'Finance',
        'Finance'
    ]
}

email_df = pd.DataFrame(emails)
email_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(email_df['message'])
y = email_df['department']

model_email = MultinomialNB()
model_email.fit(X, y)

test = vectorizer.transform(['order delayed'])
print("Predicted Department:", model_email.predict(test))
```

# 3. Fraud Detection System

```python
# Dummy fraud dataset
fraud_data = {
    'TransactionAmount': [100, 2000, 150, 5000, 300, 7000, 120, 9000],
    'LocationRisk': [1, 5, 1, 5, 2, 5, 1, 5],
    'AccountAge': [5, 1, 4, 1, 3, 1, 6, 1],
    'Fraud': [0, 1, 0, 1, 0, 1, 0, 1]
}

fraud_df = pd.DataFrame(fraud_data)
fraud_df
```

```python
X = fraud_df.drop('Fraud', axis=1)
y = fraud_df['Fraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

fraud_model = RandomForestClassifier()
fraud_model.fit(X_train, y_train)

print("Fraud Detection Accuracy:", fraud_model.score(X_test, y_test))
```
