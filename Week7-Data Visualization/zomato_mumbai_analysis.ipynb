# Zomato Mumbai Data Analysis Notebook
This notebook loads, cleans, and analyzes the dataset.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('zomato_mumbai_cleaned.csv')
df.head()
```

```python
# Dataset info
df.info()
```

```python
# Basic analysis
df['CUSINE TYPE'].value_counts().plot(kind='bar')
plt.title('Cuisine Type Distribution')
plt.show()
```

```python
# Top rated restaurants
top = df.sort_values('RATING', ascending=False).head(10)
top
```
